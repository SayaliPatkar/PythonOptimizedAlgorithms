### CRUDE APPROACH
find zeros and put them in back, shift other elements in-place
    
    1. Iterate over array
    2. if element is zero and is not block if zeros, shift next elements in place and add zero from behind
    2. if element is a block if zeros identify the complete block, shift next elements in place and add block of zeros from behind
    3. count replaced zeros and run the loop only till idx < len_arr-count

Due to nested structure (on loop for array iteration, other loop for detecting block of zero) time complexity is O(n^2)

````python
def moveZeroes(nums) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        len_arr = len(nums)
        count = 0
        for idx, elem in enumerate(nums):
            if idx == len_arr-1:
                break
            elif elem == 0 and nums[idx+1] !=0  and idx < len_arr-count:
                nums[idx:len_arr-1] = nums[idx+1:len_arr]
                nums[len_arr-1] = 0
                count += 1
            elif elem == 0 and idx < len_arr-count:
                secondary_count = 0
                secondary_idx = idx
                while(nums[secondary_idx] == 0 and secondary_idx < len_arr-1):
                    secondary_count += 1
                    secondary_idx += 1
                if idx + secondary_count == len_arr:
                    break
                nums[idx : len_arr-secondary_count] = nums[idx+secondary_count : len_arr]
                nums[len_arr-secondary_count : len_arr] = [0 for i in range(secondary_count)]
                count += secondary_count
        return nums

%time ans = moveZeroes(arr)
````

### MORE ELEGANT SOLUTION
A two-pointer approach : The idea would be to have one pointer for iterating the array and another pointer that just works on the non-zero elements of the array.

    1. initiate pointer_1 to 0
    2. Iterate over array looking for non zeros, this would be pointer_2 which is updated continuously (for loop)
    3. if element is non-zero : swap pointer_1 and pointer_2 elemets and increment pointer_1
    4. if element is a block if zeros identify the complete block, shift next elements in place and add block of zeros from behind
  

Single for loop time complexity is O(n)

````python
def moveZeroes(nums) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        len_arr = len(nums)
        idx_low = 0
        for idx in range(len_arr):
            if nums[idx] != 0:
                nums[idx_low], nums[idx] = nums[idx], nums[idx_low]
                idx_low += 1
        return nums

%time ans = moveZeroes(arr)
````
